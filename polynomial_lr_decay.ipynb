```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('ggplot')
plt.rc('xtick',labelsize=16)
plt.rc('ytick',labelsize=16)
%config InlineBackend.figure_format = 'retina'

def adjust_learning_rate(base_lr, end_lr, step, decay_steps, power=0.9):
    lr = ((base_lr - end_lr) * (1 - step / decay_steps)**(power)) + end_lr
    return lr 
```

```python
base_lr = 1e-2
end_lr = 1e-5
power = 0.9
epochs = 484
steps_per_epoch = 247
decay_steps = steps_per_epoch * epochs

lr_arr = np.zeros((epochs,))
for epoch in range(0, epochs):
    for step in range(0, steps_per_epoch):
        curr_step = epoch*steps_per_epoch + step
        lr_arr[epoch] = adjust_learning_rate(base_lr, end_lr, curr_step, decay_steps, power)
    
plt.figure(figsize=(15,8))
plt.title("Learning Schedule", fontsize=20)
plt.plot(np.arange(0,epochs), lr_arr)
plt.xlabel("Step", fontsize=18)
plt.ylabel("LR", fontsize=18)
plt.yscale("log")
plt.show()
```
